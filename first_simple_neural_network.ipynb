
## Accepting multiple weights

```python

## WEIGHTS
weights = [0.1, 0.2, 0]
## INPUTS
toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]


inputs = [toes[0], wlrec[0], nfans[0]]

def w_sum(a,b):
   
    
    assert(len(a) == len(b))
    output = 0
    for i in range(len(a)):
        output += (a[i] * b[i])
    return output
    

def neural_network(inputs, weights):
    pred = w_sum(inputs,weights)
    return pred

neural_network(inputs,weights)
```

```python
def elementwise_addition(vect_a, vect_b):
    assert(len(vect_a) == len(vect_b))
    output = [0] * len(vect_a)
    print(output)
    for i in range(len(vect_a)):
        output[i] = vect_a[i] + vect_b[i]
    return output

def elementwise_multiplication(vect_a, vect_b):
    assert(len(vect_a) == len(vect_b))
    output = [0] * len(vect_a)
    print(output)
    for i in range(len(vect_a)):
        output[i] = vect_a[i] * vect_b[i]
    return output

def vector_sum(vect_a):
    output = 0
    for i in range(len(vect_a)):
        output += vect_a[i]
    return output

def vector_average(vect_a):
    sum = vector_sum(vect_a) / len(vect_a)
    return sum


toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]


print(vector_average(toes))
```

## Using Numpy
Numpy makes it a lot easier to do dot products on an array

```python
import numpy as np

## WEIGHTS
weights = np.array([0.1, 0.2, 0])
## INPUTS
toes = np.array([8.5, 9.5, 9.9, 9.0])
wlrec = np.array([0.65, 0.8, 0.8, 0.9])
nfans = np.array([1.2, 1.3, 0.5, 1.0])

def neural_network(inputs, weights):
    pred = weights.dot(inputs)
    return pred

inputs = np.array([toes[0], wlrec[0] , nfans[0]])



print(neural_network(inputs,weights))


    
```

## Distributing single data to predict multiple outputs

```python
weight = [0.3, 0.2, 0.9]
def predict_sad_hurt_and_all(inputs, weights):
    output = [0] * len(weights)
    for i in range(len(output)):
        output[i] = inputs * weights[i]
    return output

def using_numpy(inputs,weights):
    np_weights = np.array(weights)
    return np_weights * inputs
    
wlrec = [0.65, 0.8, 0.8, 0.9]
inputs = wlrec[0]
print(using_numpy(inputs, weight))
print(predict_sad_hurt_and_all(inputs, weights))
```

```python
#           toes win fans
weights = [[0.1, 0.1,-0.3], # hurt 
           [0.1, 0.2, 0.0], # win 
           [0.0, 1.3, 0.1]] # sad

toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65,0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]


inputs = [toes[0],wlrec[0],nfans[0]]

def w_sum(a,b):
    assert(len(a) == len(b))
    output = 0
    for i in range(len(a)):
        output += (a[i] * b[i])
    return output

def vect_mat_mul(vector , matrix):
    assert(len(vector) == len(matrix))
    output = [0,0,0]
    for i in range(len(vector)):
        
        output[i] = w_sum(vector, matrix[i])
    return output


def neural_network(inputs, weights):
    pred = vect_mat_mul(inputs,weights)
    return pred


print(neural_network(inputs,weights))
```
